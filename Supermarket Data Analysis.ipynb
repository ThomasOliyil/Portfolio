```python
locationlist=sales.groupby('Location')
```

```python
location=[x for x,y in locationlist]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(location,sales.groupby('Location').sum()['Total'])
#Graphical representation of locations and its total sale
```

```python
plt.pie(sales.groupby('Location').sum()['Total'],labels=location, autopct='%1.1f%%')
#Ploting the percentage share of sales per locations
```

```python
#locations with more female customer and location with more male customer
gender_location = sales.groupby(['Gender','Location']).count()['Member']
```

```python
unstacked_location = gender_location.unstack(level=0) 
unstacked_location
```

```python
unstacked_location.plot(kind='bar')
```

```python
#Days with highest sale
sales_day= pd.to_datetime(sales['Date']).dt.day
```

```python
sales['Day'] = sales_day
```

```python
sales.head(10)
```

```python
sales.groupby('Day').sum()['Total']
```

```python
day_sale=sales.groupby('Day').sum()['Total']
```

```python
day_sale.plot()
```

```python
#branches with highest members
members= sales.groupby(['Member','Location']).count()['Invoice ID']
```

```python
members.unstack(level=0).plot(kind='bar')
```

```python
#Branchs with highest rating
rating=sales.groupby('Location').mean()['Rating']
```

```python
rating.plot(kind='bar',label='Location')
```

```python
#City with highest female shopping customer
sales.groupby(['Gender','City',]).count()['Invoice ID'].unstack().plot(kind='bar')
```

```python
#comparison of spending between men and women
sales.groupby('Gender').sum()['Total'].plot(kind='bar')
```

```python
#Which kind of customer spend more member or non_member?
sales.groupby('Member').sum()['Total'].plot(kind='bar')
```

```python
#Product line sells more?
sales.groupby('Category').sum()['Quantity'].plot(kind='bar')
```

```python
#product line most famous among male and female 
sales.groupby(['Gender','Category']).sum()['Quantity'].unstack().plot(kind='bar')
```
